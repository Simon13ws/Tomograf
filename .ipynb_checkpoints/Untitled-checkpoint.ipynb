```python
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt
import math
def discrete_radon_transform(image, steps):
    R = np.zeros((steps, len(image)), dtype='float64')
    for s in range(steps):
        rotation = misc.imrotate(image, -s*180/steps).astype('float64')
        R[:,s] = sum(rotation)
    return R

# Read image as 64bit float gray scale
image = misc.imread('Shepp_logan.jpg', flatten=True).astype('float64')
radon = discrete_radon_transform(image, 1024)

# Plot the original and the radon transformed image
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2), plt.imshow(radon, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import math 
from bresenham import bresenham
rozpietosc = math.pi/2 
alfa = math.pi
dAlfa =0/180*math.pi #zrobi input zamiast zera
liczbaDetektorow = 20
punkty = [[0,0] for i in range(liczbaDetektorow)]
image = misc.imread('Shepp_logan.jpg', flatten=True).astype('float64')
r = math.floor(math.sqrt(len(image)**2 + len(image[0])**2)/2) #okrg opisany na obrazie (nie wiem czy nie powinien by wpisany))#brakuje poozenia emitera xe i ye
square = [[1 for i in range(r*2)] for j in range(r*2)]

xe = round(r*math.cos(alfa+dAlfa))
ye = round(r*math.sin(alfa+dAlfa))
for i in range(liczbaDetektorow):
    x = r*math.cos(alfa + math.pi -(1/2)*rozpietosc + i*(rozpietosc/(liczbaDetektorow-1))+dAlfa)
    y = r*math.sin(alfa + math.pi -(1/2)*rozpietosc + i*(rozpietosc/(liczbaDetektorow-1))+dAlfa)
    punkty[i] = [round(x),round(y)]
    punkty_linii = list(bresenham(xe,ye,punkty[i][0],punkty[i][1]))
    #print(punkty_linii[724][0]+724)
    for xp,yp in punkty_linii:
        square[xp+r-1][yp+r-1] = 0
#list(bresenham(xe,ye,punkty[1][0],punkty[1][1]))


plt.figure(figsize=(30,30))
plt.imshow(square, cmap='gray')
plt.gca().invert_yaxis()
plt.show()
```
